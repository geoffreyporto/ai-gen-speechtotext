### 3. Transcribir audio .mp3 de archivo .mp4 de videollamada

```python
import subprocess

# Ruta del archivo .mp3
mp3_file = audios_path + "/Inteligencia Artificial para Indeminizaciones_new.mp3"
# Ruta del archivo de videollamada .mp4 de salida
mp4_file = videos_path + "/Inteligencia Artificial para Indeminizaciones.mp4"

# Comando para extraer el audio de .mp4 a .mp3 usando ffmpeg
subprocess.run(['ffmpeg', '-i', mp4_file, '-q:a', '0', '-map', 'a', mp3_file])

#Puedes usar la terminal para extraer el audio
#sudo ffmpeg -i "Inteligencia Artificial para Indeminizaciones.mp4" -q:a 0 -map a "Inteligencia Artificial para Indeminizaciones.mp4.mp3"


print("Audio extraído y guardado en:", mp3_file)

```

## 4. Extraer el texto del audio
Podemos realizar la conversión utilizando la herramienta `Whisper` en Python en nuesttro propio entorno local.

```python
import whisper

# Cargar el modelo de Whisper
model = whisper.load_model("large-v3")  # Puedes usar 'tiny', 'base', 'small', 'medium', 'large'
#"openai/whisper-large-v3"
# Ruta del archivo de audio .mp3
audio_file_path = audios_path + "/Inteligencia Artificial para Indeminizaciones.mp3"

# Transcribir el archivo de audio
result = model.transcribe(audio_file_path, language="es")

# Imprimir el resultado de la transcripción
print(result['text'])


# Guardar el resultado de la transcripción en un archivo de texto
output_text_path = audios_path + "/transcripcion_inteligencia_artificial_indeminizaciones_large-v3.txt"

with open(output_text_path, "w", encoding="utf-8") as text_file:
    text_file.write(result['text'])

print(f"Transcripción guardada en: {output_text_path}")
```
